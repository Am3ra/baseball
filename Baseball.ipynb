```python
## NORM
mean, var = stats.distributions.norm.fit(flat)
x = np.linspace(0,120,100)
print("norm",mean,var)

fitted_data = stats.distributions.norm.pdf(x, mean, var)


plt.hist(flat, density=True)

plt.plot(x,fitted_data,'r-')
err = np.sum(np.square(np.subtract(flat,stats.distributions.norm.pdf(flat, mean,var))))
print("norm err:",err)


## EXP
mean, var = stats.distributions.expon.fit(flat)
fitted_data = stats.distributions.expon.pdf(x, mean, var)
plt.plot(x,fitted_data,'b-')
err = np.sum(np.square(np.subtract(flat,stats.distributions.pareto.pdf(flat, mean,var))))
print("exp",mean,var)
print("exp err:",err)


## Pareto

mean, var, ex= stats.distributions.pareto.fit(flat)
fitted_data = stats.distributions.pareto.pdf(x, mean, var,ex)
plt.plot(x,fitted_data,'k-')
err = np.sum(np.square(np.subtract(flat,stats.distributions.pareto.pdf(flat, mean,var))))
print("pareto",mean,var,ex)
print("pareto err:",err)

## Lognorm

mean, var, ex= stats.distributions.lognorm.fit(flat)
fitted_data = stats.distributions.lognorm.pdf(x, mean, var,ex)
plt.plot(x,fitted_data,'c_')
err = np.sum(np.square(np.subtract(flat,stats.distributions.lognorm.pdf(flat, mean,var,ex))))
print("lognorm",mean,var,ex)
print("lognorm err:",err)


## gamma
mean, var, ex= stats.distributions.gamma.fit(flat)
fitted_data = stats.distributions.gamma.pdf(x, mean, var,ex)
plt.plot(x,fitted_data,'b-')
err = np.sum(np.square(np.subtract(flat,stats.distributions.gamma.pdf(flat, mean,var,ex))))
print("gamma",mean,var,ex)
print("gamma err:",err)


## Weib
a,c,log,scale= stats.distributions.exponweib.fit(flat)
fitted_data = stats.distributions.exponweib.pdf(x, a,c,log,scale)
plt.plot(x,fitted_data,'g-')
err = np.sum(np.square(np.subtract(flat,stats.distributions.exponweib.pdf(flat, a,c,log,scale))))
print("Weib",a,c,log,scale)
print("Weib err:",err)

```

```python
from pybaseball import statcast

# get all statcast data for July 4th, 2017
data = statcast('2017-07-04')

#get data for the first seven days of August in 2016
data = statcast('2016-07-01', '2016-08-01')
```

```python
copy = data.copy()
copy = copy[copy['launch_speed'].notna()]
copy = copy[copy['hit_distance_sc'].notna()]
copy = copy[copy['launch_angle'].notna()]
copy = copy[copy['release_speed'].notna()]
# # copy = copy[copy['spin_dir'].notna()]

target = copy.pop('hit_distance_sc')
input_data = copy[['launch_angle','launch_speed','release_speed']]
# input_data
copy
```

```python
dataset = tf.data.Dataset.from_tensor_slices((input_data.values, target.values))

for feat, targ in dataset.take(5):
  print('Features: {}, Target: {}'.format(feat, targ))
```

```python
tf.constant(copy['launch_angle'])
```

```python
train_dataset = dataset.shuffle(len(copy)).batch(1024)
```

```python
def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(140, activation='relu'),
    tf.keras.layers.Dense(140, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(40, activation='relu'),
    tf.keras.layers.Dense(40, activation='relu'),
    tf.keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error"),
                metrics=['accuracy'])
  return model

def get_compiled_model2():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error"),
                metrics=['accuracy'])
  return model
```

```python
loses = []
class LossCallback(keras.callbacks.Callback):
    def on_train_batch_end(self, batch, logs=None):
#         print("For batch {}, loss is {:7.2f}.".format(batch, logs["loss"]))
        loses.append(logs["loss"])
    
```

```python
model = get_compiled_model()
model.fit(train_dataset, epochs=40,verbose=2,callbacks=[LossCallback()])
```

```python
import matplotlib.pyplot as plt

plt.plot(loses)
```

```python
model.summary()
```

```python
from pybaseball import statcast
import pybaseball as pyball
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pybaseball.retrosheet import *

# get all statcast data for July 4th, 2017
# data = statcast('2017-07-04')

#get data for the first seven days of August in 2016
data = statcast('2016-08-01', '2016-08-08')
```

```python
import tensorflow as tf
import tensorflow.keras as keras
```

```python
# data.columns.tolist()
```

```python
cols=['p_throws','type','balls','strikes','release_pos_x','release_pos_y','pitch_type','release_speed','pfx_x','pfx_z','release_spin_rate','effective_speed','pitch_number','at_bat_number']
# cols=['p_throws']
```

```python
copy = data.copy()
copy = copy[cols]
copy = copy.dropna()
# copy = copy[copy['p_throws'].notna()]
# len(copy['type'].notna())
copy['p_throws'] = pd.Categorical(copy['p_throws'])
copy['p_throws']=copy.p_throws.cat.codes
# copy['type'] = pd.Categorical(copy['type'])
# copy['type']=copy.type.cat.codes
# copy = copy.assign(type = lambda x: if x['type'] == 'X' return 1 else return 0)
copy['pitch_type'] = pd.Categorical(copy['pitch_type'])
copy['pitch_type']=copy.pitch_type.cat.codes
copy['type'] = [1 if t == 'X' else 0 for t in copy['type']]
target = copy.pop('type')
# copy.head()
target
```

```python
def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(140, activation='relu'),
    tf.keras.layers.Dense(140, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(40, activation='relu'),
    tf.keras.layers.Dense(40, activation='relu'),
    tf.keras.layers.Softmax()
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error"),
                metrics=['accuracy'])
  return model

def get_compiled_model2():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error"),
                metrics=['accuracy'])
  return model
```

```python
loses = []
class LossCallback(keras.callbacks.Callback):
    def on_train_batch_end(self, batch, logs=None):
#         print("For batch {}, loss is {:7.2f}.".format(batch, logs["loss"]))
        loses.append(logs["loss"])
```

```python
dataset = tf.data.Dataset.from_tensor_slices((copy.values, target.values))
train_dataset = dataset.shuffle(len(copy)).batch(1024)
```

```python
loses = []
model = get_compiled_model()
model.fit(train_dataset, epochs=200,verbose=0,callbacks=[LossCallback()])
```
