```python
data = pybaseball.schedule_and_record(2019, "HOU").columns
data = pd.DataFrame(columns=data)


def getTeamSchedules(initial_year,final_year):
    teamdata = []
    
    for team in bbref_current_teams:
        df = pybaseball.schedule_and_record(initial_year, team)
        for year in range(initial_year+1,final_year+1):
            df = df.append(pybaseball.schedule_and_record(year, team))
        teamdata.append(df)
    return teamdata
```
